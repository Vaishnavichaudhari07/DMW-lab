```python
# Find the missing values in Each Column:
data.isnull().sum()
```

```python
# Find the missing values from Rating Column and repalce by Median value
# user Defined function
def impute_median(series):
    return series.fillna(series.median())

# Apply user defined function into Rating Column:
data.work_year=data['work_year'].transform(impute_median)
```

```python
data.isnull().sum()
```

```python
# convet price and review and rating into numerical Column
data['salary']=data['salary'].apply(lambda x: str(x).replace('$','')if '$' in str(x) else str(x))
```

```python
data['salary']=data['salary'].apply(lambda x: float(x))
```

```python
data.describe()
```

```python
data['salary_in_usd']=pd.to_numeric(data['salary_in_usd'],errors='coerce')
```

```python
data['remote_ratio']=data['remote_ratio'].apply(lambda x: str(x).replace('+','')if '+' in str(x) else str(x))
```

```python
data['remote_ratio']=data['remote_ratio'].apply(lambda x: str(x).replace(',','')if ',' in str(x) else str(x))
data['remote_ratio']=data['remote_ratio'].apply(lambda x: float(x))
```
